### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Either by accessing `pyplot` directly, or by the object-oriented method of creating `Figure`- and `Axes`-objects and calling their methods.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

**Figure**

A `Figure` is an object that holds the `Axes` and `Artists` objects.

**Axes**

The `Axes` object contains the `Axis` objects of the `Figure`.

**Axis**

An `Axis` contains the data points, labels and such of the plot.

**Artist**

The `Artist` object is responsible for actually visualizing the plot to the screen or to an image file.

### When plotting in Matplotlib, what is the expected input data type?

Numbers, either as floats or integers.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4)
y = x**2

# Pyplot approach
plt.plot(*[x, y])
plt.title('A plot, pyplot-style')
plt.xlabel('x')
plt.ylabel('y');

# OO-approach
fig, ax = plt.subplots()
ax.plot(*[x, y])
ax.set(title='A plot, OO-style', xlabel='x', ylabel='y')
```
